```python
for column in data.columns:
    if column != "diagnosis":
        plt.figure(figsize=(6, 4))
        sns.kdeplot(data= data, x=column, hue="diagnosis", fill=True, common_norm=False, palette=["green", "red"])
        plt.title(f"Density Plot for {column}")
        plt.xlabel(column)
        plt.ylabel("Density")
        plt.legend(["Benign (0)", "Malignant (1)"])
        plt.tight_layout()
        plt.show()
```

```python
#BETTER WAY OF DISPLAY
fig, axes = plt.subplots(5, 6, figsize=(20, 15))
axes = axes.flatten()

i=0
for column in data.columns:
    if column != "diagnosis":
        sns.kdeplot(
            data=data,
            x=column,
            hue="diagnosis",
            fill=True,
            common_norm=False,
            palette=["green", "red"],
            ax=axes[i]
        )
        axes[i].set_title(f"Density Plot for {column}")
        axes[i].set_xlabel("")
        axes[i].set_ylabel("")
        axes[i].legend([], [], frameon=False)
        i=i+1

plt.tight_layout()
plt.show()
```

```python
X = data.drop('diagnosis', axis=1)
Y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train , Y_test = train_test_split(X,Y, test_size=0.2 , )

#Why random state = 7 ... thala for a reason...
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression(class_weight='balanced')
lr.fit(X_train, Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import classification_report

print("Classification Report:")
print(classification_report(Y_test, Y_pred, target_names=["Benign", "Malignant"]))
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {acc:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
acc=accuracy_score(Y_test, Y_pred)*100
print(f"Accuracy: {acc} %")
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("Classification Report:\n", classification_report(Y_test, Y_pred, target_names=["Benign", "Malignant"]))
```

```python
grid = GridSearchCV(estimator=xgb, param_grid=param_grid, 
                    cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

grid.fit(X_train, Y_train)
```
